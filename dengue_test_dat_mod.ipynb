```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
test_df = pd.read_csv("dengue_features_test.csv")
test_df.head()
```

```python
test_df.isnull().sum()
```

```python
tt_df = test_df.drop(["week_start_date", "ndvi_ne","ndvi_nw","ndvi_se","ndvi_sw", "precipitation_amt_mm","reanalysis_avg_temp_k","reanalysis_dew_point_temp_k","reanalysis_min_air_temp_k","reanalysis_precip_amt_kg_per_m2","reanalysis_relative_humidity_percent","reanalysis_relative_humidity_percent","reanalysis_sat_precip_amt_mm","reanalysis_specific_humidity_g_per_kg","reanalysis_tdtr_k","station_avg_temp_c","station_diur_temp_rng_c","station_max_temp_c","station_min_temp_c","station_precip_mm"], axis=1)
tt_df.head()
```

```python
# dist from monsoon
tt_df["d_monsoon"] = ""
for i in range(tt_df.shape[0]):
  if tt_df.iloc[i,0] == "iq":
    tt_df.iloc[i,5] = abs(tt_df.iloc[i,2] - 10)
  else:
    tt_df.iloc[i,5] = abs(tt_df.iloc[i,2] - 42)

tt_df = tt_df.drop(["city","weekofyear"], axis=1)
tt_df
```

```python
# scaling all columns except year

col_scal = ["reanalysis_air_temp_k", "reanalysis_max_air_temp_k", "d_monsoon"]
scaler = StandardScaler()
tt_df1 = scaler.fit_transform(tt_df[col_scal])
```

```python
tt_df1 = pd.DataFrame(data = tt_df1, columns = col_scal)
tt_df1
```

```python
tt_df2 = pd.concat([tt_df1, tt_df["year"]], axis = 1)
tt_df2.head()
```

```python
tt_df2.isnull().sum()
```

```python
tt_df2["reanalysis_air_temp_k"].fillna(value=tt_df2["reanalysis_air_temp_k"].mean(), inplace = True)
tt_df2["reanalysis_max_air_temp_k"].fillna(value=tt_df2["reanalysis_max_air_temp_k"].mean(), inplace = True)
```

```python
tt_df2.to_csv("mod_test_dat.csv")
```
